```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('playstore.csv')
```

```python
df.head()
```

### 1) All the apps comes under ART_AND_DESIGN

```python
df[df['category'] == 'ART_AND_DESIGN']
```

### 2) Print the number of paid and free apps

```python
print("Free : ", len(df[df['type'] == 'Free']))
print("Paid : ", len(df) - len(df[df['type'] == 'Free']))
```

### 3) List number of apps in each category

```python
for name in df['category'].unique():
    print(name ," : ", len(df[df['category'] == name]))
```

### 4) Top 5 categories in terms of number of applicaions

```python
data = []

for name in df['category'].unique():
    data.append([name , len(df[df['category'] == name])])
    
df_pr = pd.DataFrame(data, columns = ['name','freq'])
```

```python
df_pr.sort_values(by = 'freq', ascending = False).head()
```

### 5) Top 5 content_rating in terms of number of application

```python
data = []

for name in df['content_rating'].unique():
    data.append([name , len(df[df['content_rating'] == name])])
    
df_pr = pd.DataFrame(data, columns = ['content_rating','freq'])

df_pr.sort_values(by = 'freq' , ascending = False).head()
```

### 6) List the unrated apps in content_rating

```python
df[df['content_rating'] == 'Unrated']
```

### 7) Apps in FAMILY with rating more than 4.5

```python
df_pr = df[df['category'] == 'FAMILY']
df_pr = df_pr[df_pr['rating'] > 4.5]
```

```python
df_pr
```

### 8) Average Price apps in each content rating

```python
for name in df['content_rating'].unique():
    print(name, int(df[df['content_rating'] == name]['price'].mean()))
```

### 9) Maximum/Minimum Price apps in each content rating

```python
for name in df['content_rating'].unique():
    print(name, df[df['content_rating'] == name]['price'].max(), df[df['content_rating'] == name]['price'].min())
```

### 10) Average number of reviews in each category

```python
for name in df['category'].unique():
    print(name, int(df[df['category'] == name]['review'].mean()))
```

### 11) Top 5 categories in terms of total number of reviews

```python
data = []

for name in df['category'].unique(): 
    data.append([name , df[df['category'] == name]['review'].sum()])
    
pd.DataFrame(data, columns = ['category','review']).sort_values(by = 'review', ascending = False).head()
```

### 12) Number of apps and number of reviews ratio in content_rating

```python
for name in df['content_rating'].unique():
    print(name , int(df[df['content_rating'] == name]['review'].sum() / len(df[df['content_rating'] == name])))
```

### 13) TOP 5 number of apps and number of reviews ratio in category

```python
data = []

for name in df['category'].unique():
    data.append([name , int(df[df['category'] == name]['review'].sum() / len(df[df['category'] == name]))])
    
df_pr = pd.DataFrame(data, columns = ['category','ratio'])
```

```python
df_pr.sort_values(by = 'ratio',ascending = False).head()
```

### 14) List the paid apps whose category in FAMILY, Content Rating is everyone and rating == 5 

```python
df_pr = df[df['category'] == 'FAMILY']
df_pr = df_pr[df_pr['type'] == 'Paid']
df_pr = df_pr[df_pr['content_rating'] == 'Everyone']
df_pr = df_pr[df_pr['rating'] == 5]
```
