### 1.2) Top 10 apps based on their Reviews

```python
df.head()
```

```python
apps = set()

for i in df.sort_values(by = 'Reviews', ascending = False).values:
    apps.add(i[0])    
    if (len(apps) == 10):
        break
```

```python
c = 1
for i in apps:
    print(c,":",i)
    c+=1
```

### 1.3) Top 10 apps based on their Reviews (Sorted Format based on Apps)

```python
lst = []

for app in apps:   
    for i in df.values:       
        if (i[0] == app):
            lst.append([i[0],i[3]])
            break
            
df_pr = pd.DataFrame(lst, columns = ['name','reviews'])

c = 1

for i in df_pr.sort_values(by = 'reviews', ascending = False).values:
    print(c, ":", i[0])
    c += 1
```

### 1.4) Create a new dataframe with no duplicate app name

```python
df.head()
```

```python
df.shape
```

```python
dct = {}

for i in df.values:    
    dct[i[0]] = {'category' : i[1], 'rating' : i[2], 
                 'review' : i[3] ,'size' : i[4], 
                 'installs' : i[5], 'type' : i[6],
                 'price' : i[7], 'content_rating' : i[8]}
```

```python
lst = []

for name in dct.keys():
    lst.append([name, dct[name]['category'],dct[name]['rating'],dct[name]['review'],
          dct[name]['size'], dct[name]['type'], dct[name]['installs'], dct[name]['price'],
          dct[name]['content_rating']])
```

```python
df_pr = pd.DataFrame(lst, columns = ['app','category','rating','review','size','type','installs','price','content_rating'])
```
