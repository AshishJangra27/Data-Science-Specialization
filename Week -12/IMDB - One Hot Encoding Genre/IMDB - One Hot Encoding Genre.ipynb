```python
import numpy as np
import pandas as pd

df = pd.read_csv('imdb.csv')
df.head()
```

#### 1.1) Removing Columns that are of no use

```python
del df['Poster_Link']
del df['Released_Year']
del df['Certificate']
del df['Runtime']
del df['Overview']
del df['Meta_score']
del df['Director']
del df['Star1']
del df['Star2']
del df['Star3']
del df['Star4']
del df['No_of_Votes']
del df['Gross']
del df['IMDB_Rating']

df.head()
```

#### 1.2) Checking Null Values

```python
df.isnull().sum()
```

#### 1.3) Finding All Genre

```python
genre = []
for i in df['Genre']:
    
    genre += i.split(',')
    
genre = [i.strip() for i in genre]    
genre = list(set(genre))
```

#### 1.4) One Hot Conversion

```python
one_hot = []

for i in df['Genre']:
    
    i = [j.strip() for j in i.split(',')]        # Removing Spaces
    enc = [0 for i in range(21)]
    
    for gen in i:
        enc[genre.index(gen)] = 1
    
    one_hot.append(enc)
```

#### 1.5) One Hot Conversion Verification (Counting Number of Genres and 1's)

```python
for i in range(len(df['Genre'])):
    
    x1 = len(df['Genre'][i].split(','))          # Finding Number of Genres in the Movie
    x2 = list(one_hot[i]).count(1)               # Finding Number of 1's in One Hot List
    
    if (x1 != x2):
        print('Not Equal')
```

#### 1.6) Converting OneHot to Array

```python
one_hot = np.array(one_hot)

one_hot.shape
```

#### 1.7) Adding OneHotData Data to DataFrame

```python
for i in range(21):
    df[genre[i]] = one_hot[ : ,i]
    
df.head()
```
