```python
import matplotlib.pyplot as plt
```

# 1. Pie Chart

### Level - 1

```python
marks = [45,50,55,60]
plt.pie(marks)
plt.show()
```

### Level - 2 | Add Labels

```python
marks  = [45,50,55,60]
labels = ['Hindi','English','Math','Science']

plt.pie(marks,labels = labels)
plt.show()
```

### Level - 3 | Adding Angles

```python
marks  = [45,50,55,60]
labels = ['Hindi','English','Math','Science']

plt.pie(marks, labels = labels, startangle = 103)
plt.show()
```

### Level - 4 | Adding Explode

```python
marks  = [45,50,55,60]
labels = ['Hindi','English','Math','Science']

explodes = [.01,.01,.01,.1]

plt.pie(marks, labels= labels, explode = explodes)
plt.show()
```

### Level - 5 | Adding Shadows

```python
marks  = [45,50,55,60]
labels = ['Hindi','English','Math','Science']

explodes = [.01,.01,.01,.1]

plt.pie(marks, labels= labels, explode = explodes, shadow = True)
plt.show()
```

### Level - 6 | Adding Specific Colours

```python
marks  = [45,50,55,60]
labels = ['Hindi','English','Math','Science']
colours = ['red','#989304','g','orange']

explodes = [.01,.01,.01,.1]

plt.pie(marks, labels= labels, explode = explodes, shadow = True, colors = colours)
plt.show()
```

### Level - 7 | Add Legend with Labels

```python
marks  = [45,50,55,60]
labels = ['Hindi','English','Math','Science']

plt.pie(marks, labels= labels, explode = explodes, shadow = True)
plt.legend()
plt.show()
```

### Level - 8 | Adding Only Legend

```python
marks  = [45,50,55,60]

plt.pie(marks, explode = explodes, shadow = True)
plt.legend(['Hindi','English','Math','Science'])
plt.show()
```

### Level - 9 | Adding Title

```python
marks  = [45,50,55,60]

plt.pie(marks, explode = explodes, shadow = True)
plt.legend(['Hindi','English','Math','Science'], title = 'Subjects')
plt.show()
```

### Level - 10 | Adding Numbers

```python
marks  = [45,50,55,60]
labels = ['Hindi','English','Math','Science']

plt.pie(marks, explode = explodes, shadow = True, labels = labels, autopct = '%1.2f%%')
plt.show()
```

# 2. Scatter Plot

### Level - 1 | Plot for Girls

```python
girls = [88,90,91,89,95]

standard = [6,7,8,9,10]

plt.scatter(standard, girls, edgecolors = 'pink' , marker = 's', c = 'pink' , linewidths = 2)

plt.legend(['Girls'])

plt.xlabel('Class')
plt.ylabel('Marks')

plt.title('Marks Distribution of Girls in different classes')

plt.show()
```

### Level - 2 | Plot for Boys

```python
boys  = [78,87,88,83,90]


standard = [6,7,8,9,10]

plt.scatter(standard, boys, edgecolors = 'blue' , marker = '^', c = 'pink' , linewidths = 2)

plt.legend(['Boys'])

plt.xlabel('Class')
plt.ylabel('Marks')

plt.title('Marks Distribution of Boys in different classes')

plt.show()
```

### Level - 3 | Combining Boys and Girls

```python
girls = [88,90,91,89,95]
boys  = [78,87,88,83,90]

standard = [6,7,8,9,10]

plt.scatter(standard, girls, edgecolors = 'pink' , marker = 's', c = 'pink' , linewidths = 2)
plt.scatter(standard, boys , edgecolors = 'blue', marker = '^', c = 'blue', linewidths = 2)

plt.legend(['Girls','Boys'])

plt.xlabel('Class')
plt.ylabel('Marks')

plt.title('Marks Distribution of boys and Girls in different classes')

plt.show()
```

### Level - 4 | Adding Min/Max

```python
girls = [88,90,91,89,95]
boys  = [78,87,88,83,90]

minimum = [0,0,0,0,0]
maximum = [100,100,100,100,100]

standard = [6,7,8,9,10]

plt.scatter(standard, girls, edgecolors = 'pink' , marker = 's', c = 'pink' , linewidths = 2)
plt.scatter(standard, boys , edgecolors = 'blue', marker = '^', c = 'blue', linewidths = 2)

plt.scatter(standard, minimum, c = 'white')
plt.scatter(standard, maximum, c = 'white')

plt.legend(['Girls','Boys'])

plt.xlabel('Class')
plt.ylabel('Marks')

plt.title('Marks Distribution of boys and Girls in different classes')

plt.show()
```
