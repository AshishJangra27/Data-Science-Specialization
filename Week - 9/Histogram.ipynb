```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Level - 1 (Basic Plot with small data)

```python
marks = [65,77,45,56,87,68,98,80,34,54,56,76,77,45,65,78,76,88,49,94]

fig, ax = plt.subplots()
ax.hist(marks, bins = 10)

plt.xlabel('Marks')
plt.ylabel('Frequency')

plt.title('Marks Distribution')

plt.show()
```

### Level - 2 (Plots with large data) - Random

```python
data = []
for i in range(10000):
    data.append(np.random.random())
    
fig, axs = plt.subplots(figsize = (10,5))
axs.hist(data, bins = 100)
plt.show()
```

```python
data = []
for i in range(10000):
    data.append(np.random.random() - .5)
    
fig, axs = plt.subplots(figsize = (10,5))
axs.hist(data, bins = 100)
plt.show()
```

### Level - 3 (Normal Distribution)

```python
data = np.random.randn(1000000)

fig , axs = plt.subplots(figsize = (6,4))

axs.hist(data, bins = 1000)

plt.title('Normal Distribution')

plt.show()
```

### Level - 4 (Normal Distribution in Specific Range)

```python
data = np.random.randn(1000000)

data = (data + 4)*12.5

fig , axs = plt.subplots(figsize = (6,4))

axs.hist(data, bins = 1000)

plt.title('Normal Distribution')

plt.show()
```

### Level - 5 (Normal Distribution on Google Playstore Dataset)

```python
df = pd.read_csv('playstore.csv')
df.head()
```

```python
data = list(df['rating'])

fig , axs = plt.subplots()
axs.hist(data, bins = 30)

plt.show()
```

### Level - 6 (Data Distribution of Ratings) - Scratch

```python
data = list(df['rating'])

ratings = []

for rating in set(data):
    c = 0
    for i in range(len(data)):
        if (data[i] == rating):
            c+= 1
            
    ratings.append([rating,c])
    
df_rating = pd.DataFrame(ratings, columns = ['rating','freq'])

rating = list(df_rating.sort_values(by = 'rating')['rating'])
freq   = list(df_rating.sort_values(by = 'rating')['freq'])
```

```python
plt.bar(rating, freq)
plt.show()
```

### Level - 7 (Making it to Normaly Distributed) [rating > 3.6]

```python
data = []

for i in df['rating']:
    if (i > 3.6):
        data.append(i)


fig , axs = plt.subplots()
axs.hist(data, bins = [3.7,3.8,3.9,4.0,4.1,4.2,4.3,4.4,4.5,4.6,4.7,4.8,4.9,5.0])

plt.show()
```

### Level - 8 (Making it to Normaly Distributed) [rating > 3.6 and rating < 5]

```python
data = []

for i in df['rating']:
    if (i > 3.6 and i < 5.0):
        data.append(i)


fig , axs = plt.subplots()
axs.hist(data, bins = 13)

plt.show()
```

### Level - 8 (Making it to Normaly Distributed) Finalising Rating 4.4

```python
final_data = []

c = 0

for i in range(len(data)):
    
    if (data[i] == 4.4 and c <= 1200):
        c += 1
        final_data.append(data[i])
        
    elif (data[i] == 4.4):
        pass
        
    else:
        final_data.append(data[i])
```

```python
fig , axs = plt.subplots()
axs.hist(final_data, bins = 13)

plt.show()
```
