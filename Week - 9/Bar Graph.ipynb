### Level - 1 (Default Setting)

```python
plt.bar(student,marks)
plt.show()
```

### Level - 2 (Width Settings)

```python
plt.bar(student,marks, width = .5)

plt.show()
```

### Level - 3 (Color Settings)

```python
plt.bar(student,marks, color = 'green')

plt.show()
```

### Level - 4 (Title Settings)

```python
plt.bar(student,marks, color = 'blue')

plt.xlabel('Names')
plt.ylabel('Marks')

plt.title('Comparing Student Marks')

plt.show()
```

### Level - 7 (Stacked Graphs)

### Level - 8 (Boys to Girls Ratio)

```python
class_ = ['First', 'Second', 'Third', 'Fourth', 'Firth', 'Sixth', 'Seventh','Eigth', 'Ninth', 'Tenth']
boys   = [40     , 32      ,  23    ,      43 ,  33    , 44     , 24       , 25    , 43     , 22     ] 
girls  = [21     , 23      ,  23    ,      12 ,  11    , 21     , 30       , 21    , 16     , 15     ] 

f1 = []
f2 = []

for i in range(len(boys)):
    
    f1.append(int(girls[i] / (boys[i]+girls[i])*100))
    f2.append(int(girls[i] / (boys[i])*100))
    
plt.bar(class_, f2)
```

### Level - 9 (Advance Stacked Graph)

```python
name = ['Ashish', 'Raj', 'Rohal', 'Ruhul']

eng = np.array([46,56,76,78])
hin = np.array([88,56,76,88])
mth = np.array([73,63,67,87])
sci = np.array([88,67,87,88])
```

```python
plt.bar(name,sci, color = 'g')                                     # Science - Green
plt.bar(name,mth, bottom = sci, color = 'b')                       # Maths   - Blue
plt.bar(name,hin, bottom = sci+mth, color = 'r')                   # Hindi   - Red
plt.bar(name,eng, bottom = sci+mth+hin , color = 'orange')         # English - Orange

plt.title('Comparing student marks in different subjects')
plt.xlabel('Names')
plt.ylabel('Marks')

plt.legend(['Science','Math','Hindi','English'])

plt.show()
```
