## 1. Dataset Preprocessing
### 1.1) Dataset Loading

```python
df = pd.read_csv('/kaggle/input/ted-talk-youtube/TED.csv')

df.head()
```

### 1.2) Dealing with Null Values

```python
df.isnull().sum()
```

```python
del df['thumbnail_link']
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

## 2) Data Preprocessing

```python
df.head()
```

### 2.1) Preprocessing Views Column

```python
views = []

for i in df['views']:
    views.append(int(i.replace(',','')))
    
df['views'] = views
```

```python
df.head()
```

### 2.2) Preprocessing likes column

```python
likes = []

for i in df['likes']:
    
    if (i == 'Like'):
        likes.append('Like')
        
    elif ('K' in i):
        likes.append(int(float(i[:-1])*1000))
    
    elif ('M' in i):
        likes.append(int(float(i[:-1])*1000000))
        
    else:
        likes.append(int(i))
        
df['likes'] = likes
```

```python
df.head()
```

### 2.3) Preprocessing Comments Column

```python
comments = []

for i in df['comments']:
    comments.append(int(i.replace(',','')))
    
df['comments'] = comments 
```

```python
df.head()
```

### 2.4) Preprocessing Date Column

```python
date    = []
month  = []
year   = []


for i in df['date']:
    if ('Streamed' in i):
        
        date.append(int(i.split(' ')[-3:][0]))
        month.append(i.split(' ')[-3:][1])
        year.append(int(i.split(' ')[-3:][2]))
        
    else:
        date.append(int(i.split(' ')[0]))
        month.append(i.split(' ')[1])
        year.append(int(i.split(' ')[2]))

df['date']   = date
df['month'] = month
df['year']  = year
```

```python
df.head()
```

```python
del df['description']
```

### 2.5) Preprocessing Month Column

```python
df.head()
```

```python
dct = {'Jan' : 1, 'Feb' : 2, 'Mar' : 3, 'Apr' : 4, 'May' : 5 , 'Jun' : 6,
       'Jul' : 7, 'Aug' : 8, 'Sept' : 9 , 'Oct' : 10, 'Nov' : 11 , 'Dec' : 12}

month = []

for i in df['month']:
    month.append(dct[i])
    
df['month'] = month
```

```python
df.head()
```

### 2.6) Combining Day,Month and Year

```python
date = []

for i in df.values[:,-3:]:
    date.append(str(i[0]) + '/' + str(i[1]) + '/'  + str(i[2]))
    
df['date_month_year'] = date
```

```python
df.head()
```

### 2.7) Filling Likes with View-to-Like Ration

```python
df.head()
```

```python
ratio = []

for i in df.values[:,3:5]:
    
    if (i[1] != 'Like'):
        ratio.append(i[0]/i[1])
        
ratio = int(int(sum(ratio))/len(ratio))
```

```python
likes = []

for i in df.values[:,3:5]:
    
    if (i[1] == 'Like'):
        likes.append(int(i[0]/72))
    else:
        likes.append(i[1])
        
df['likes'] = likes
```

```python
df.head()
```

### 2.8) Finding Day of publishing of every video

```python
day = []

for i in df['date_month_year']:
    day.append(datetime.date(int(i.split('/')[2]),int(i.split('/')[1]),int(i.split('/')[0])).strftime('%a'))
    
df['day'] = day
```

```python
df.head()
```

### 2.9) Preprocessing duration column

```python
def duration(lst): 
    minutes = 0    
    for i in lst.split(','):         
        if ('minutes' in i):               # Checking Minutes   (As it is)
            minutes += int(i[:-8])
        if ('hour' in i):                  # Checking Hours     (Multiply by 60)
            minutes += (int(i[:-5]) * 60)
        if ('seconds' in i):               # Checking Seconds   (Increment by 1 if seconds are more than 40)
            if (int(i[:-8]) > 40):
                minutes += 1 
    return minutes

dur = []
for i in df['duration']:
    dur.append(duration(i))
    
df['duration'] = dur
```
