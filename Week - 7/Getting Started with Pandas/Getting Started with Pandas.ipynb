## Creating a dataframe

```python
df = pd.read_csv('Data.csv', header = None)

name = df[0]
eng = df[1]
hin = df[2]
mth = df[3]
sci = df[4]

del df[0]
del df[1]
del df[2]
del df[3]
del df[4]
```

```python
df['name'] = name
df['eng'] = eng
df['hin'] = hin
df['mth'] = mth
df['sci'] = sci
```
