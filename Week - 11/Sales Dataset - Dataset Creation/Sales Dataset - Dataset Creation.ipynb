```python
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

### 1) Timezone

```python
df_1 = df[df['timezone'] == 'AM']
df_2 = df[df['timezone'] == 'PM']

df_1.to_csv('Timezone/AM.csv', index = False)
df_2.to_csv('Timezone/PM.csv', index = False)
```

### 2) State

```python
!mkdir State

for i in df['state'].unique():
    path = 'State/'+i+".csv"
    
    df_ = df[df['state'] == i]
    df_.to_csv(path, index = False)
```

### 3) City | Adding total Number of Products Sold

```python
data = []

for city in df['city'].unique():
    
    q = 0
    o = 0
    a = 0
    
    
    for i in df.values:
        if (i[12] == city):          # i[12] is City
            o += 1                   # Count the Order
            q += i[2]                # i[2] is Quantity, Count the Quantity
            a += i[4]                # i[4] is Total, Count the total amount

    data.append([city,o,q,int(a)])
    
df_ = pd.DataFrame(data, columns = ['City', 'Orders','Quantity','Amount'])

df_.to_csv('City.csv',index = False)
```

### 4) City

```python
!mkdir City

for i in df['city'].unique():
    
    path = 'City/' + i + '.csv'
    df_ = df[df['city'] == i]
    df_.to_csv(path, index = False)
```

### 5) Products

```python
data = []

for i in df['product'].unique():
    
    df_ = df[df['product'] == i]
    

    data.append([i, 
                 df_['price'].mode().values[0], 
                 len(df_), 
                 df_['quantity'].sum(), 
                 df_['quantity'].max(), 
                 int(df_['total'].sum())])
    
df_ = pd.DataFrame(data , columns = ['Name','Price','Orders','Quanitity','Max_quantity_per_user', 'Amount'])

df_.to_csv('Product.csv', index = False)
```
